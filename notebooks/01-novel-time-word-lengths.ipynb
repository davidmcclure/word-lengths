```python
import matplotlib as mpl
import matplotlib.pyplot as plt

%matplotlib inline
mpl.style.use('bmh')
```

```python
from word_length import Text

wp = Text.from_file('../data/2600-0.txt')
```

```python
lengths = wp.token_lengths()

plt.figure(figsize=(20, 5))
plt.scatter(range(5000), lengths[:5000])
plt.show()
```

```python
import numpy as np
import statsmodels.api as sm

y = wp.token_lengths()
X = np.arange(len(y))

result = sm.OLS(y, X).fit()
```
